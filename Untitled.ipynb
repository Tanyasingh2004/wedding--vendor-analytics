```python
import pandas as pd
vendors = pd.read_csv("vendors.csv")
services = pd.read_csv("services.csv")
bookings = pd.read_csv("bookings.csv")
```

```python
vendors.head()
vendors.info()
vendors.isnull().sum()
```

```python
vendors.drop_duplicates(inplace=True)
```

```python
vendors.columns = vendors.columns.str.lower().str.strip()
```

```python
vendors = vendors[(vendors['rating']>= 1)&(vendors['rating']<=5)]
```

```python
vendors.describe()
```

```python
services.drop_duplicates(inplace=True)
services['service_price'] = pd.to_numeric(services['service_price'],errors='coerce')
```

```python
services = services[services['service_price']>1000]
```

```python
bookings['booking_date'] = pd.to_datetime(bookings['booking_date'])
```

```python
bookings = bookings[bookings['price_paid']>0]
```

```python
bookings.fillna({'event_type':"Wedding"},inplace=True)
```

```python
bookings = bookings[bookings['vendor_id'].isin(vendors['vendor_id'])]
services = services[services['vendor_id'].isin(vendors['vendor_id'])]
```

```python
final_data = bookings.merge(vendors,on='vendor_id')
final_data = final_data.merge(services,on='vendor_id')
final_data.head()
```

```python
final_data.to_csv("final_wedding_data.csv",index=False)
```

```python
import pandas as pd
df = pd.read_csv("final_wedding_data.csv")
```

```python
x = df[['city','category_x','rating']]
y =df['price_paid']
```

```python
from sklearn.preprocessing import LabelEncoder
le_city = LabelEncoder()
le_category_x = LabelEncoder()
x['city'] = le_city.fit_transform(x['city'])
x['category_x'] = le_category_x.fit_transform(x['category_x'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)

print(prediction[:5])
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, prediction)

print(mae)
```

```python
import joblib

joblib.dump(model, "wedding_model.pkl")
```
